```python
from configPy import DirManager, EnvManager, Config
import pandas as pd

file_dir = Config.get_dir_files()
data_file = file_dir.get_any("dados.xlsx")
```

## Separação da planilha em multiplos arquivos csv

```python
# diretório de saída das planilhas
planilhas_dir = file_dir.create_dir("planilhas")

# arquivo carregado

excel_file = pd.read_excel(data_file, sheet_name=None)

# iterar em cima das planilhas
for nome_da_aba, dados in excel_file.items():
    name_csv = planilhas_dir.create_file_path(f"planilha_{nome_da_aba}", suffix="csv")
    dados.to_csv(name_csv, index=False, encoding='utf-8')
    print(f"Arquivo '{name_csv}' criado com sucesso")
```

Baseado na estrutura dessa planilha é possível perceber que é um registro detalhado de execução financeira (gastos reais).

Principais pontos:

**1. Estrutura hierarquica de contas**

Os dados seguem um Plano de Contas estruturado em três níveis de detalhamento

- 1 Nível: Grande categoria (ex: despesa) divide entre Despesa e receita
- 2 Nível: Grupo intermediário (ex: Despesa corrente, Receita Correntes) especifica o tipo de receita o despesa
- 3 Nível: Grupo expecífico (ex: Pessoal e Encargos, Ocupação e Utilidades) especifica o tipo de atividade da despesa/receita

Cada uma dessas categorias possui um código associado.

**2. Segmentação por Centro de Custo**

A coluna "Centro de Responsabilidade" (ex: "Mirante") indica que a empresa ou instituição divide seus gastos por unidades, departamentos ou projetos específicos. Isso é fundamental para identificar quais áreas estão consumindo mais recursos.

**3. Natureza temporal**

Os dados são registrados mensalmente ("Mês" e "nº mês") e anualmente ("Ano").

As linhas mostradas referem-se a Janeiro de 2022.

Essa estrutura permite criar análises de sazonalidade e evolução temporal (comparar um mês com o outro).

**4. Métrica Principal: "Realizado"**

A coluna final, "Realizado", contém os valores monetários que de fato foram gastos. Diferente de uma coluna de "Orçado" (planejado), o "Realizado" mostra o que já aconteceu.
